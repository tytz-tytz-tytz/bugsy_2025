```python
import sys
from pathlib import Path
import os
os.chdir("..")

from prompts_with_results.prompt_builder import build_prompt, save_prompt_to_file
from semantic.search import semantic_search
from graph_search.search_engine import normalize_query_to_stems, graph_search_full
from graph_search.semantic_expander import expand_tags
```

```python
query = (
    "Редактирование пользовательского события"
)

keywords = [
    "start", "блок", "сообщение", "параметры", "удаление", 
    "отмена", "форма", "интент", "сценарий", "дублирование"
]
```

```python
# Пример: функция поиска с keyword boosting
context_chunks = semantic_search(query=query, top_k=10, keywords=keywords)

# Текстовые данные
context_texts = [c["text"] for c in context_chunks if "text" in c]

for i, text in enumerate(context_texts, 1):
    print(f"\nФрагмент {i}:\n{text[:300]}...")
```

```python
scenario = (
    "Редактирование пользовательского события"
)

prompt_text = build_prompt(
    context_chunks=context_texts,
    scenario=scenario,
)
```
