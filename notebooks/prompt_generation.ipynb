```python
from src.retriever.search_smart import search_smart_chunks

# Запрос и ключевые слова, которые повышают приоритет чанков
query = (
    "Конструктор для задания сценария рассылки. "
    "Интерфейс настройки рассылки. "
    "Блоки действий. Связи между блоками. Изменение сценария."
)

# Ключевые слова, усиливающие релевантность
keywords = [
    "start", "блок", "сообщение", "конфиденциальные данные",
    "перевод в канал", "отмена сообщений", "отмена работы кнопок",
    "вызов внешней системы", "генерация события", "генерация интента",
    "SMS", "форма", "связь", "дублирование", "браузер"
]

context_chunks = search_smart_chunks(query=query, k=10, keywords=keywords)

for i, chunk in enumerate(context_chunks, 1):
    print(f"\nФрагмент {i}:")
    print(chunk.get("text"))
```

```python
from src.generator.prompt_builder_v2 import build_prompt_no_dom

scenario = "Протестируй работу конструктора сценариев в интерфейсе Marketer. Ограничься только сценарием конструктора рассылки: работа с блоками действий, настройка параметров, взаимодействие между блоками. Не включай тесты, связанные с SMS-провайдерами, лендингами, API, выгрузками, deeplink и т.д."

context_texts = [c["text"] for c in context_chunks if "text" in c]

# Генерация промпта без DOM
prompt = build_prompt_no_dom(
    context_chunks=context_texts,
    scenario=scenario
)

# Сохраняем
with open("../data/results/generated_prompt_rag_nodom.txt", "w", encoding="utf-8") as f:
    f.write(prompt)

print("Промпт без DOM сохранён в: ../data/results/generated_prompt_UX_3.txt")
```
